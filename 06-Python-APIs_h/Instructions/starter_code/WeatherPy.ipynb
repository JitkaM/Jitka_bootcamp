```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
api_key = api_keys.api_key
#print(api_key)

max_temperature = []
humidity = []
cloudiness = []
wind_speed = []
latitude = []
city_list = []
country = []
date = []
longitude = []
record_counter = 1
set_counter = 1

#import json

print("Beginning Data Retrieval")
print("-----------------------------")

for city in cities:
    try:
        query_url = url + "appid=" + api_key + "&q=" + city + "&units=imperial"
        weather_response_json = requests.get(query_url).json() 
        #print(json.dumps(weather_response_json, indent=4, sort_keys=True))
        city_list.append(weather_response_json["name"])
        max_temperature.append(weather_response_json["main"]["temp_max"])
        humidity.append(weather_response_json["main"]["humidity"])
        cloudiness.append(weather_response_json["clouds"]["all"])
        wind_speed.append(weather_response_json["wind"]["speed"])
        latitude.append(weather_response_json["coord"]["lat"])
        longitude.append(weather_response_json["coord"]["lon"])
        country.append(weather_response_json["sys"]["country"])
        date.append(weather_response_json["dt"])
        #print(set_counter)
        #set_counter += 1
        if record_counter % 51 == 0:
            set_counter += 1
            #print(set_counter)
            record_counter = 1
        print("Processing Record " + str(record_counter) + " of Set " + str(set_counter) + " | " + weather_response_json["name"])
        
        record_counter += 1
    except KeyError:
        print("City not found. Skipping...")
    except TypeError:
        print("TypeError. Skipping...")

print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
#for city in city_list:
    #print(city)
#for temp in max_temperature:
    #print(temp)
#for humi in humidity:
    #print(humi)
#for cloud in cloudiness:
    #print(cloud)
#for wind in wind_speed:
    #print(wind)
#for lat in latitude:
    #print(lat)
#for lon in longitude:
    #print(lon)
#for c in country:
    #print(c)
#for dt in date:
    #print(dt)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df = pd.DataFrame({'City':city_list, 'Cloudiness':cloudiness, 'Country':country, 'Date': date, 'Humidity':humidity, 'Lat': latitude, 'Lng': longitude, 'Max Temp': max_temperature, 'Wind Speed': wind_speed})
#weather_df.count()
weather_df.to_csv("jitka_cities_weather.csv", sep='\t')
weather_df.count()
```
